### Q11

```python
np.random.seed(2)
X = np.random.gamma(1, 5, 1000)
Y = np.random.poisson(lam=5, size=1000)
```

- (a) Histogram Plots\
**The gamma distribution X is not normally distributed** but **the poisson distribution Y is normal** upon visual inspection.

```python
_,_,_ = plt.hist(X, bins=15)
plt.xlabel('Gamma Distribution X')
plt.ylabel('Frequency')
plt.title('Histogram of Gamma Distribution X')
plt.show()

_,_,_ = plt.hist(Y, bins=15)
plt.xlabel('Poisson Distribution Y')
plt.ylabel('Frequency')
plt.title('Histogram of Poisson Distribution Y')
plt.show()
```

- (b) ECDF Plots\
**The gamma distribution X is not visually normal** but **poisson distribution Y is visually and approximately normal** compared to the ECDF plot of the theoretical CDF functions from normal distribution.

```python
from statsmodels.distributions.empirical_distribution import ECDF

mean_X = np.mean(X)
std_X = np.std(X)

mean_Y = np.mean(Y)
std_Y = np.std(Y)

norm_X = np.random.normal(mean_X, std_X, size=1000)
norm_Y = np.random.normal(mean_Y, std_Y, size=1000)

normX = ECDF(norm_X)
normY = ECDF(norm_Y)

ecdfX = ECDF(X)
ecdfY = ECDF(Y)
_ = plt.plot(ecdfX.x, ecdfX.y)
_ = plt.plot(normX.x, normX.y)
plt.xlabel('Gamma Distribution X')
plt.ylabel('ECDF')
plt.legend(['sampled data', 'normal distribution'])
plt.title('ECDF of Gamma Distribution X')
plt.show()
_ = plt.plot(ecdfY.x, ecdfY.y)
_ = plt.plot(normY.x, normY.y)
plt.xlabel('Poisson Distribution Y')
plt.ylabel('ECDF')
plt.legend(['sampled data', 'normal distribution'])
plt.title('ECDF of Poisson Distribution Y')
plt.show()
```

- (c) Statistical Tests\
I plan to use two-tailed **unpaired t-Test (parametric)** and **Mann-Whitney U-Test (non-parametric)** to evaluate whether or not X is significant distinct from Y. As far as I am thinking, **the non-parametric Mann-Whitney U-Test** is well suited to distinguish X from Y since distribution X and Y do not match the normality assumption and we cannot use the parametric test to evaluate the statistics between X and Y.

```python
alpha = 0.05
_, un_p_value = ttest_ind(X, Y, alternative='two-sided')
if un_p_value < alpha:
    print("The unpaired t-Test suggests that X and Y are statistically distinct.")
else:
    print("The unpaired t-Test does not provide evidence that X and Y are statistically distinct.")
    
_, p_value = mannwhitneyu(X, Y, alternative='two-sided')
if p_value < alpha:
    print("The Mann-Whitney U-Test suggests that X and Y are statistically distinct.")
else:
    print("The Mann-Whitney U-Test does not provide evidence that X and Y are statistically distinct.")
```

### Q14

- (a) PCA Model\
I prefer to use the multivariate normality test for joint distribution and Shapiro-Wilk test for the marginal distribution of first two components. **The joint distribution of the first two components is not normally distributed and the marginal distribution of the first two components is not normally distributed.**

```python
pca_model = PCA(n_components=10).fit(db)
components = pca_model.transform(db)
plt.figure(figsize=(5, 5))
plt.scatter(components[:,0], components[:, 1])
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('First Two Principal Components')
plt.show()

# joint distribution of the first two principal components
_,pvalue,_ = multivariate_normality(components[:,:2])
if pvalue > 0.05:
    print('The joint distribution of the first two components is approximately normally distributed.')
else:
    print('The joint distribution of the first two components is not normally distributed.')

# marginal distribution of the first two principal components
_, pvalue1 = shapiro(components[:,0])
_, pvalue2 = shapiro(components[:,1])
if pvalue1 > 0.05 and pvalue2 > 0.05:
    print('The marginal distribution of the first two components is approximately normally distributed.')
else:
    print('The marginal distribution of the first two components is not normally distributed.')
```

- (b) Plot Explained Variance\
**The number of components that are needed to capture 90% of the variance is 7**

```python
v = pca_model.explained_variance_
fig, ax = plt.subplots(1,2, figsize=(12,5))
ax[0].plot(np.arange(len(v))+1, v)
ax[1].plot(np.arange(len(v))+1, np.cumsum(v))
ax[0].set_xticks(np.arange(len(v))+1)
ax[1].set_xticks(np.arange(len(v))+1)
ax[0].set_xlabel('Principal Component Number')
ax[1].set_xlabel('Principal Component Number')
ax[0].set_ylabel('Explained Variance')
ax[1].set_ylabel('Cumulative Explained Variance')
ax[0].set_title('Explained Variance')
ax[1].set_title('Cumulative Explained Variance')
plt.show()
variance_ratio = pca_model.explained_variance_ratio_
print(f'The number of components that are needed to capture 90% of the variance: {np.where(np.cumsum(variance_ratio)>=0.9)[0][0]+1}')
```

- (a) Plot Image and Histogram

```python
import numpy as np
import scipy.stats

def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h

fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].imshow(img, cmap='gray')
ax[1].hist(img.flatten())
ax[0].set_title('Test Image')
ax[1].set_title('Histogram of Pixel Values of Test Image')
ax[1].set_xlabel('Pixel Value')
ax[1].set_ylabel('Frequency')
plt.show()
interval = np.percentile(img.flatten(), [2.5, 97.5])
print(f'The 95% confidence interval of the pixel values of test image is {interval}')
```

- (b) Apply Sobel Filter Along X Axis

```python
Dx = sobel(img, axis=1)
plt.imshow(Dx, cmap='gray')
plt.title("Sobel Filter on X Axis")
plt.show()
```

- (c) Apply Sobel Filter Along Y Axis

```python
Dy = sobel(img, axis=0)
plt.imshow(Dy, cmap='gray')
plt.title("Sobel Filter on Y Axis")
plt.show()
```

- (d) Plot Dxy
    - Similarity
        - All three images show the edge extraction of the original image.
        - Dx and Dy are the components of the Dxy along the x and y axes respectively. Dxy has the similar edge detection as Dx and Dy.
    - Differences
        - Dx and Dy are the edge detection along different axes, while Dxy captures all the edges of the original image.
        - The edges detected by Dx and Dy may have different in both magnitude and direction, but the edges in Dxy are only the positive magnitudes.

```python
Dxy = np.sqrt(Dx ** 2 + Dy ** 2)
plt.imshow(Dxy, cmap='gray')
plt.title('Image Dxy')
plt.show()
```

- (e) Compare Dxy and Test Image\
The histogram of Dxy is not the same as the histogram of Test Image.
    - Dxy gives the changes of gradient in both x and y directions or the intensity changes in the image. Dxy is mainly influenced by the intensity changes of the image, which results in difference compared to the pixel values of Test Image.
    - From the histogram statistics of Dxy and Test Image, Dxy gives higher values of skewness, kurtosis and variance but lower values of the mean values. This means that Dxy has narrow pixel range, large pixel spread and long heavy tails which represents the characteristic of sharp edges in the Dxy image compared to the test image.


```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].hist(img.flatten())
ax[0].set_title('Test Image')
ax[0].set_title('Histogram of Pixel Values of Test Image')
ax[0].set_xlabel('Pixel Value')
ax[0].set_ylabel('Frequency')
ax[1].hist(Dxy.flatten())
ax[1].set_title('Image Dxy')
ax[1].set_title('Histogram of Pixel Values of Image Dxy')
ax[1].set_xlabel('Pixel Value')
ax[1].set_ylabel('Frequency')
plt.show()

# Histogram stats for each image
img_stats = describe(img.flatten())
Dxy_stats = describe(Dxy.flatten())

print("Image Histogram Stats:")
print("Mean:", img_stats.mean)
print("Variance:", img_stats.variance)
print("Skewness:", img_stats.skewness)
print("Kurtosis:", img_stats.kurtosis)

print("\nDxy Histogram Stats:")
print("Mean:", Dxy_stats.mean)
print("Variance:", Dxy_stats.variance)
print("Skewness:", Dxy_stats.skewness)
print("Kurtosis:", Dxy_stats.kurtosis)

plt.show()
```
